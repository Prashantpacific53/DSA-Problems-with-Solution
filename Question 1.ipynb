# Q1: Find the Maximum in a Subarray (Sliding Window) 
Reference: LeetCode #643 (variant)

Given an integer array nums and a window size k, slide the window from left to right and print the
maximum element in each window.

Example: nums = [1, 3, -1, -3, 5, 3], k = 3 → [3, 3, 5, 5]

Hint: Use a simple loop. For each window position i, scan nums[i] to nums[i+k-1] and track the max. Start with
brute force O(nk), then think about how to avoid rescanning.

Tags: Array · Sliding Window · Brute Force First

# 🧠 Q1: Maximum in Each Subarray of Size k
📌 Problem (in simple words)

You are given an array and a window size k.

👉 Move a window of size k from left → right

👉 For every window, find the maximum element

✅ Example

nums = [1, 3, -1, -3, 5, 3]

k = 3

Windows:

[1, 3, -1] → 3

[3, -1, -3] → 3

[-1, -3, 5] → 5

[-3, 5, 3] → 5

👉 Output: [3, 3, 5, 5]

# Step 1: Brute Force Approach (IMPORTANT for understanding)
## 💡 Idea:

- For every window → scan all k elements → find max

# 🧮 Time Complexity:
Total windows = n - k + 1

Each window takes k time

O(n × k)

```python
def maxSlidingWindow(nums, k):
    result = []
    
    for i in range(len(nums) - k + 1):
        max_val = nums[i]
        
        for j in range(i, i + k):
            if nums[j] > max_val:
                max_val = nums[j]
        
        result.append(max_val)
    
    return result

# Example
nums = [1, 3, -1, -3, 5, 3]
k = 3
print(maxSlidingWindow(nums, k))  # [3, 3, 5, 5]
```

# 🔹 Step 2: Dry Run (VERY IMPORTANT 🧠)

Take first window:

i = 0 → [1, 3, -1]

max = 3

Next:

i = 1 → [3, -1, -3]

max = 3

Repeat → done ✅


🔥 Problem with Brute Force

We are recalculating max again and again ❌

👉 Example:

3 was already max in previous window

But we still recompute 😑

# 🔹 Step 3: Optimization Thinking (Core Concept)

👉 Ask yourself:

“Can I reuse previous computation?”

YES ✅

Use a Deque (Double-ended Queue)

## 🚀 Optimized Approach (Sliding Window + Deque)

💡 Idea:

Maintain a deque storing indices of useful elements

Rules:

Remove elements out of window

Remove smaller elements (they can't be max)

Front of deque = max element

## 🧮 Time Complexity:

👉 O(n) (each element processed once)

```python
from collections import deque

def maxSlidingWindow(nums, k):
    dq = deque()
    result = []
    
    for i in range(len(nums)):
        
        # Remove elements out of window
        if dq and dq[0] == i - k:
            dq.popleft()
        
        # Remove smaller elements
        while dq and nums[dq[-1]] < nums[i]:
            dq.pop()
        
        dq.append(i)
        
        # Store result when window is valid
        if i >= k - 1:
            result.append(nums[dq[0]])
    
    return result

# Example
nums = [1, 3, -1, -3, 5, 3]
k = 3
print(maxSlidingWindow(nums, k))  # [3, 3, 5, 5]
```

# 🔥 Intuition (REMEMBER THIS FOREVER)

## 👉 Deque always stores:

Elements in decreasing order

👉 Why?

Smaller elements are useless ❌

Bigger element will dominate future windows ✅

## 🧠 Memory Trick

Think like this:

“Only strong elements survive in the queue 😎”

Weak (smaller) elements → removed
Old elements → removed

Only potential max remains

## 🎯 When to Use This Pattern

Use this whenever:

Fixed window size k

Need max / min / sum / condition per window
