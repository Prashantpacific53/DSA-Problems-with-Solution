# 🧠 Q2: Reverse a Linked List
## 📌 1. First: What is a Linked List?
💡 Definition:

A Linked List is a collection of nodes where each node contains:data a pointer/reference to the next node

## 📦 Structure:
[1 | next] → [2 | next] → [3 | next] → None

## 🧠 Key Terms:
Head → first node
Tail → last node (points to None)
Next pointer → connects nodes

## 🔄 2. What does “Reverse a Linked List” mean?
💡 Simple Meaning:

Change direction of links

Before:

1 → 2 → 3 → 4 → 5 → None

After:

5 → 4 → 3 → 2 → 1 → None

👉 We are NOT changing values

👉 We are changing links (pointers)

## 🔥 3. Approach 1: Iterative (Most Important)
💡 Core Idea:

We reverse links one by one using 3 pointers

## 🧠 Important Pointers:

Pointer	Meaning

prev	previous node

curr	current node

next	next node (to save link)

## 🔁 Algorithm Steps:
Store next node

Reverse link

Move pointers forward

## 🔄 Step-by-Step Visualization

Initial:


prev = None

curr = 1 → 2 → 3 → 4 → 5

Step 1:

next = 2

1 → None

prev = 1

curr = 2

Step 2:
next = 3

2 → 1

prev = 2

curr = 3

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def reverseList(head):
    prev = None
    curr = head
    
    while curr:
        next_node = curr.next   # Step 1: save next
        curr.next = prev        # Step 2: reverse link
        prev = curr             # Step 3: move prev
        curr = next_node        # Step 4: move curr
    
    return prev
```

```python
def reverseList(head):
    # Base case
    if not head or not head.next:
        return head
    
    new_head = reverseList(head.next)
    
    head.next.next = head   # reverse link
    head.next = None        # break old link
    
    return new_head
```

```python
# Definition for singly-linked list node
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


# -------------------------------
# Helper function to create linked list from list
# -------------------------------
def create_linked_list(arr):
    if not arr:
        return None
    
    head = ListNode(arr[0])
    current = head
    
    for value in arr[1:]:
        current.next = ListNode(value)
        current = current.next
    
    return head


# -------------------------------
# Helper function to print linked list
# -------------------------------
def print_linked_list(head):
    current = head
    while current:
        print(current.val, end=" -> ")
        current = current.next
    print("None")


# =====================================================
# 🔁 1. ITERATIVE APPROACH
# =====================================================
def reverse_iterative(head):
    prev = None        # previous node
    curr = head        # current node
    
    while curr:
        next_node = curr.next   # Step 1: store next node
        curr.next = prev        # Step 2: reverse link
        prev = curr             # Step 3: move prev forward
        curr = next_node        # Step 4: move curr forward
    
    return prev   # new head


# =====================================================
# 🔁 2. RECURSIVE APPROACH
# =====================================================
def reverse_recursive(head):
    # Base case
    if head is None or head.next is None:
        return head
    
    # Reverse rest of the list
    new_head = reverse_recursive(head.next)
    
    # Reverse current node
    head.next.next = head
    head.next = None
    
    return new_head


# =====================================================
# 🔥 DRIVER CODE (TESTING)
# =====================================================
if __name__ == "__main__":
    arr = [1, 2, 3, 4, 5]
    
    # Create linked list
    head = create_linked_list(arr)
    
    print("Original Linked List:")
    print_linked_list(head)
    
    # Iterative Reverse
    reversed_head_iter = reverse_iterative(head)
    print("\nReversed (Iterative):")
    print_linked_list(reversed_head_iter)
    
    # Reverse again using recursive (to get original back)
    reversed_head_rec = reverse_recursive(reversed_head_iter)
    print("\nReversed Again (Recursive):")
    print_linked_list(reversed_head_rec)
```

🧠 Key Concept (VERY IMPORTANT)

👉 Why do we store next?

Because:

curr.next = prev

will break the original link ❌

So we must save it first ✅

🧠 Memory Trick:

“Save → Reverse → Move” 🔁

🚀 4. Approach 2: Recursive
💡 What is Recursion?
Definition:

A function that calls itself

🧠 Core Idea:
Reverse the rest of the list
Fix the first node at the end
🔁 Thinking:
1 → 2 → 3 → 4 → 5

👉 First reverse:

2 → 3 → 4 → 5 → becomes → 5 → 4 → 3 → 2

👉 Then attach:

1 at the end
🧾 Code (Recursive)

🔥 Important Concept in Recursion
This line is MAGIC:
head.next.next = head

👉 Meaning:

2 → 1
And this:
head.next = None

👉 Breaks original link

🧠 Memory Trick:

“Go deep → reverse → fix on return”

⚖️ Iterative vs Recursive
Feature	Iterative	Recursive
Speed	Faster	Slightly slower
Space	O(1)	O(n) (stack)
Easy to understand	Yes	Medium
🔥 Most Important Concepts to Remember
✅ 1. Pointer Manipulation

👉 You are changing connections

✅ 2. Data Structure Behavior

Sequential

✅ 3. In-place Algorithm
